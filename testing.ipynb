```python
import jager
import requests
import bs4
```

```python
url = "http://contagiodump.blogspot.com/2014/07/cz-solution-ltd-signed-samples-of.html"
```

```python
r = requests.get(url)
```

```python
r_text = r.text
```

```python
r_text
```

```python
soup = bs4.BeautifulSoup(r_text)
```

```python
[s.extract() for s in soup('script')]
```

```python
my_text = soup.body.get_text()
print my_text
```

```python
jager.extract_domains(my_text)
```

```python
jager.extract_filenames(my_text)
```
